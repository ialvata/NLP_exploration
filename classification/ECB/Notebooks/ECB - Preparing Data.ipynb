# ECB Rates

```python
# this section will be changed in the future.
```

```python
def multiclass(value):
    if value==0:
        return 0
    if value>0:
        return 2
    if value<0:
        return 1
```

```python
df_MRO_1 = pd.read_csv(path+"\\Data - Raw\\"+"MRO_Interest_Rate.csv",skiprows=5) 
# 5 initial lines are irregular.

df_MRO_1.columns=["Dates","MRO_Rates", "Obs"]
del df_MRO_1["Obs"]
df_MRO_1['Dates']=pd.to_datetime(df_MRO_1.Dates,dayfirst=True).astype(str)
df_MRO_1=df_MRO_1.fillna(method="bfill")
df_MRO_1["Rate Difference"]=df_MRO_1["MRO_Rates"].diff(periods=1) # difference that points to evolution
df_MRO_1["Classes"]=df_MRO_1["Rate Difference"].apply(multiclass)
df_MRO_1.to_csv(path+"\\Data - Clean\\"+"MRO_Rates.csv", index = False)
```

# Consolidating Data = Speeches + Rates

```python
#df_speech_EN=df5 #df3[df3["what_language"].isin(["EN",np.NaN])]
df_speech_EN_merged=pd.merge(left=df5,right=df_MRO_1,
                             how="left",left_on=["date"],
                             right_on=["Dates"])
# to see the speeches we're losing 
# run df_speech_EN_merged[df_speech_EN_merged["MRO_Rates"].isna()==True]
# Previous to 1999, there was no Common European Currency (EUR)
```

```python
df_speech_EN_merged=df_speech_EN_merged[df_speech_EN_merged["MRO_Rates"].isna()==False]

df_speech_EN_merged.to_csv(path+"\\Data - Clean\\"+"Consolidated_Data.csv",index=False)
```
